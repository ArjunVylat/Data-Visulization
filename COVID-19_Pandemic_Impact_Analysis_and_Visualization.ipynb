Dataset Loading

```python
import pandas as pd

covid_data = pd.read_csv('all-states-history.csv')

covid_data.head()
```

Data Preprocessing

```python
covid_data.info()
```

```python
covid_data.describe()
```

```python
missing_values = covid_data.isnull().sum()
missing_values
```

```python
numerical_columns = covid_data.select_dtypes(include = ['int64', 'float64']).columns

for column in numerical_columns:
    covid_data[column].fillna(covid_data[column].mean(), inplace = True)
```

```python
covid_data.isna().sum()
```

```python
covid_data.describe()
```

```python
covid_data['date'] = pd.to_datetime(covid_data['date'])
```

```python
categorical_columns = covid_data.select_dtypes(include = ['object']).columns
categorical_columns
```

```python
covid_data['state'] = covid_data['state'].astype('category')
```

```python
covid_data.drop_duplicates(inplace = True)
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    return outliers

outliers = detect_outliers_iqr(covid_data, numerical_columns)

outliers
```

```python
import numpy as np 

def handle_outliers_transformation(data, column):
    data[column] = np.where(data[column] > 0, np.log(data[column]), data[column])

handle_outliers_transformation(covid_data, numerical_columns)


covid_data.describe()
```

Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = covid_data.corr()

plt.figure(figsize = (12, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize = (8, 5))
    sns.histplot(covid_data[column], bins = 20, kde = True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize = (8, 5))
    sns.boxplot(x = covid_data[column])
    plt.title(f'BoxPlot of {column}')
    plt.xlabel('column')
    plt.show()
```

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = covid_data, x = covid_data['state'])
plt.title('Count Plot of States')
plt.xlabel(categorical_columns)
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```
